4. Los elementos que componen la estructura de datos de Git son blob, tree y commit. Blob representa todo el contenido de un archivo en un momento dado, es un bloque de datos binarios sin metadatos. Tree representa un directorio para blobs y otros tree (subtrees), estos representan una estructura jerarquica donde existe un arbol raiz, el cual contiene los archivos y carpetas del proyecto. Commit representa un punto en el historial del proyecto, este contiene información sobre autor, el mensaje del commit, una marca de tiempo y un hash que apunta a un tree.

Por último, investigue y comente para qué sirven los siguientes comandos:

11. `git stash` (0.25 puntos)
12. `git reset`. ¿Cuál es la diferencia de este comando con `git revert`?. (0.25 puntos)

<code>Escriba sus respuestas acá:</code>

11. <code>git stash</code> se utiliza para guardar temporalmente los cambios en un directorio de trabajo sin realizar un commit. Esto es útil cuando se tienen cambios incompletos o no deseados, pero se desea cambiar de rama o aplicar actualizaciones sin comprometer los cambios actuales. git stash toma los cambios no confirmados y los guarda en una pila de almacenamiento (stash), lo que permite volver al estado limpio del último commit.

12. La diferencia entre <code>git reset</code> y <code>git revert</code> es que el primero se utiliza para mover la referencia de una rama a un commit específico mientras que el segundo se utiliza para crear un nuevo commit que deshace los cambios introducidos por un commit específico manteniendo el historial intacto.
Hay tres formas principales de utilizar <code>git reset</code>

    git reset --soft <hash del commit>: Mueve la referencia de la rama y mantiene los cambios de los commits posteriores en el área de preparación (staging area). Esto permite realizar un nuevo commit con los cambios acumulados.

    git reset --mixed <hash del commit>: Mueve la referencia de la rama, descarta los cambios en el área de preparación, pero mantiene los cambios en el directorio de trabajo. Esto permite revisar los cambios nuevamente antes de agregarlos al área de preparación y realizar un nuevo commit.

    git reset --hard <hash del commit>: Mueve la referencia de la rama y descarta tanto los cambios en el área de preparación como en el directorio de trabajo. Los cambios se perderán de manera irreversible.
