````python
def generar_descripcion(fruta):
    # Documentación Aquí
    """Genera una descripción de una fruta especificada en base a sus propiedades.

    Esta función toma un diccionario de fruta como entrada y genera una descripción
    detallada de la fruta, incluyendo su nombre, tipo, color y presencia de pepas.

    Parameters
    ----------
    fruta : dict
        Un diccionario que contiene información sobre la fruta. Debe tener las siguientes keys:
        - 'nombre' (str): El nombre de la fruta.
        - 'tipo' (str): El tipo o categoría de la fruta.
        - 'color' (str): El color de la fruta.
        - 'pepas' (bool): Un valor booleano que indica si la fruta tiene pepas (True) o no (False).

    Returns
    -------
    str
        Una descripción de la fruta en función de sus propiedades. Si la fruta no está en la base de datos,
        se devuelve un mensaje indicando que la fruta no se encuentra en los datos.

    """


    # Código Aquí
    descripcion = " "
    if fruta == None:
        return "El parámetro está vacio"
    for frutas in datos:
        if unidecode(frutas["nombre"]) == unidecode(fruta["nombre"]):
            color = frutas["color"]
            tipo = frutas["tipo"]
            if frutas["pepas"] == True:
                pepas = "Presenta pepas"
                if unidecode(frutas["nombre"]) == "limon":
                    descripcion = "El {} es una fruta de tipo {} de color {}. {} en su interior.".format(frutas["nombre"], tipo, color, pepas)
                else:
                    descripcion = 'La {} es una fruta de tipo {} de color {}. {} en su interior.'.format(frutas["nombre"], tipo, color, pepas)
            elif frutas["pepas"] == False:
                pepas = "No presenta pepas"
                if frutas["nombre"] == "plátano":
                    descripcion = "El {} es una fruta de tipo {} de color {}. {} en su interior.".format(frutas["nombre"], tipo, color, pepas)
                else:
                    descripcion = 'La {} es una fruta de tipo {} de color {}. {} en su interior.'.format(frutas["nombre"], tipo, color, pepas)
    if descripcion == " ":
        descripcion = "La fruta ingresada no se encuentra en nuestros datos"
    return descripcion
````

**2. La función `describir(datos)` que generalice la función `generar_descripcion` para recibir una lista de frutas y devolver una lista con las descripciones de cada una:**

Por ejemplo, `describir([limon, naranja]])` deberá devolver una lista de la forma `[descripcion_limon, descripcion_naranja]`

````python
def describir(datos):
    # Documentación Aquí
    """Genera descripciones para una lista de frutas en base a sus propiedades.

    Esta función toma una lista de diccionarios de frutas como entrada y genera descripciones detalladas
    para cada fruta en la lista, utilizando la función generar_descripcion.

    Parameters
    ----------
    datos : list
        Una lista que contiene diccionarios de frutas. Cada diccionario debe tener las siguientes claves:
        - 'nombre' (str): El nombre de la fruta.
        - 'tipo' (str): El tipo o categoría de la fruta.
        - 'color' (str): El color de la fruta.
        - 'pepas' (bool): Un valor booleano que indica si la fruta tiene pepas (True) o no (False).

    Returns
    -------
    list
        Una lista de descripciones de frutas generadas en función de sus propiedades.
        Si la lista de datos está vacía, se devuelve un mensaje indicando que el parámetro está vacío.

    """
    
    # Código Aquí
    if datos == None:
        return "El parámetro está vacio"
    descripciones = []
    for frutas in datos:
        descrip = generar_descripcion(frutas)
        descripciones.append(descrip)
    return descripciones
````

**3. La función `filtrar_por_pepa(datos, tiene_pepas)` que dado un tipo de fruta y un booleano, retorne las frutas que tienen o no tienen pepas (según el valor de `tiene_pepas`).**
    

Por ejemplo, `filtrar_por_pepa(datos, True)` deberá retornar una lista con los diccionarios de: naranja, limon, frutilla y frambuesa.

````python
def filtrar_por_pepa(datos, tiene_pepas):
     
    # Documentación Aquí
     """Filtra una lista de frutas en base a la presencia de pepas.

    Esta función toma una lista de diccionarios de frutas y un valor booleano que indica si se deben
    filtrar las frutas que tienen pepas (True) o las que no tienen pepas (False).

    Parameters
    ----------
    datos : list
        Una lista que contiene diccionarios de frutas. Cada diccionario debe tener las siguientes claves:
        - 'nombre' (str): El nombre de la fruta.
        - 'tipo' (str): El tipo o categoría de la fruta.
        - 'color' (str): El color de la fruta.
        - 'pepas' (bool): Un valor booleano que indica si la fruta tiene pepas (True) o no (False).
    tiene_pepas : bool
        Un valor booleano que determina si se deben filtrar las frutas con pepas (True) o sin pepas (False).

    Returns
    -------
    list
        Una lista de diccionarios de frutas filtradas en función de la presencia de pepas.
        Si la lista de datos está vacía, se devuelve un mensaje indicando que el parámetro está vacío.

    """
    # Código Aquí
     if datos == None:
          return "El parámetro está vacio"
     filtrados=[]
     for frutas in datos:
         if frutas["pepas"] == tiene_pepas:
            filtrados.append(frutas)
     return filtrados
````

**4. La función `conteo_colores(datos)` que cree un diccionario que haga un conteo los colores de las frutas.**

Para estos datos, la función debería retornar el siguiente diccionario: 
    
```python
{"amarillo": 3, "naranjo": 1, "rojo": 2}
```

````python
def conteo_colores(datos):
    # Documentación Aquí
    """Realiza un conteo de colores de frutas en la lista de datos.

    Esta función toma una lista de diccionarios de frutas y realiza un conteo de cuántas frutas tienen
    cada color en la lista.

    Parameters
    ----------
    datos : list
        Una lista que contiene diccionarios de frutas. Cada diccionario debe tener las siguientes claves:
        - 'nombre' (str): El nombre de la fruta.
        - 'tipo' (str): El tipo o categoría de la fruta.
        - 'color' (str): El color de la fruta.
        - 'pepas' (bool): Un valor booleano que indica si la fruta tiene pepas (True) o no (False).

    Returns
    -------
    dict
        Un diccionario que contiene el conteo de colores de frutas en función de sus propiedades.
        Si la lista de datos está vacía, se devuelve un mensaje indicando que el parámetro está vacío.

    """
    # Código Aquí
    if datos == None:
        return "El parámetro está vacio"
    conteo = {}
    for frutas in datos:
        color = frutas["color"]
        if color in conteo:
            conteo[color] += 1
        else:
            conteo[color] = 1         
    return conteo
````

**5. La función `obtener_tipos` que devuelva una lista con los tipos únicos de fruta usando `set()`**.

Por ejemplo, `obtener_tipos([limon, naranja, platano])` debería devolver `["citrica", "tropical"]`

````python
def obtener_tipos(datos):
    # Documentación Aquí
    """Obtiene una lista de tipos de frutas presentes en la lista de datos.

    Esta función toma una lista de diccionarios de frutas y devuelve una lista única
    de los diferentes tipos de frutas presentes en la lista.

    Parameters
    ----------
    datos : list
        Una lista que contiene diccionarios de frutas. Cada diccionario debe tener las siguientes claves:
        - 'nombre' (str): El nombre de la fruta.
        - 'tipo' (str): El tipo o categoría de la fruta.
        - 'color' (str): El color de la fruta.
        - 'pepas' (bool): Un valor booleano que indica si la fruta tiene pepas (True) o no (False).

    Returns
    -------
    list
        Una lista de tipos de frutas únicos presentes en la lista de datos.
        Si la lista de datos está vacía, se devuelve un mensaje indicando que el parámetro está vacío.

    """
    # Código Aquí
    if datos == None:
        return "El parámetro está vacio"
    tipos_con_rep = []
    for frutas in datos:
        tipo = frutas["tipo"]
        tipos_con_rep.append(tipo)
    tipos = list(set(tipos_con_rep))
    return tipos
````
